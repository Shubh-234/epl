```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
#n_estimators is the number of decision trees you wish to train
#the larger the number the longer the algorithm will take to execute but it will also be more accurate
#min_samples: no of samples we wish to take,the higher the number the least chance of overfit
```

```python
train = matches[matches["date"] < '2023-01-01']
```

```python
test = matches[matches["date"] > '2023-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
#fitting the predictors in the random forest model to predict that target is 0 or 1(lose or win)
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
def rolling_averages(group,cols,new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3,closed = "left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
#compute the stats of the previous 3 matchweeks of that team to predict how that team will perform in that upcoming match
#we use closed = ="left" because we do not want to take current matchweek data(future data) into consideration
#drop na dispose the null values encountered
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
#take the macthes dataset by team,assign a dataframe to each team 
#and apply and compute the rolling averaged on each team dataframe
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2023-01-01']
    test = data[data["date"] > '2023-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
